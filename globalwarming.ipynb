```python
import pandas as pd
import os

csv_dosyalari = [f for f in os.listdir(path) if f.endswith('.csv')]

temiz_veriler = {}

print("Data Cleaning\n")

for dosya in csv_dosyalari:
    dosya_yolu = os.path.join(path, dosya)
    df = pd.read_csv(dosya_yolu)

    onceki_satir_sayisi = len(df)

    df_temiz = df.dropna().copy()

    sonraki_satir_sayisi = len(df_temiz)
    silinen_satir = onceki_satir_sayisi - sonraki_satir_sayisi

    temiz_veriler[dosya] = df_temiz

    print(f"File: {dosya}")
    print(f"  - Total Row: {onceki_satir_sayisi}")
    print(f"  - Deleted (NaN) Row: {silinen_satir}")
    print(f"  - Remaining Row: {sonraki_satir_sayisi}")
    print("-" * 30)
```

```python
for dosya_adi, df_temiz in temiz_veriler.items():

    print(f"File: {dosya_adi} (The first 10 rows have been cleaned.)")


    display(df_temiz.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df_global = temiz_veriler['GlobalTemperatures.csv'].copy()


df_global['dt'] = pd.to_datetime(df_global['dt'])
df_global['year'] = df_global['dt'].dt.year


yearly_temp = df_global.groupby('year')['LandAverageTemperature'].mean()


plt.figure(figsize=(14, 7))
sns.set_style("whitegrid")

plt.plot(yearly_temp.index, yearly_temp.values, label='Annual Average Temperature', color='tab:red', alpha=0.4)


plt.plot(yearly_temp.index, yearly_temp.rolling(window=10).mean(), label='10-Year Trendline', color='darkred', linewidth=2.5)


plt.title('Global Land Temperature Change (1750 - 2015)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.legend()


plt.show()
```
