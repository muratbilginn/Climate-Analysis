```python
import pandas as pd
import os

csv_dosyalari = [f for f in os.listdir(path) if f.endswith('.csv')]

temiz_veriler = {}

print("Data Cleaning\n")

for dosya in csv_dosyalari:
    dosya_yolu = os.path.join(path, dosya)
    df = pd.read_csv(dosya_yolu)

    onceki_satir_sayisi = len(df)

    df_temiz = df.dropna().copy()

    sonraki_satir_sayisi = len(df_temiz)
    silinen_satir = onceki_satir_sayisi - sonraki_satir_sayisi

    temiz_veriler[dosya] = df_temiz

    print(f"File: {dosya}")
    print(f"  - Total Row: {onceki_satir_sayisi}")
    print(f"  - Deleted (NaN) Row: {silinen_satir}")
    print(f"  - Remaining Row: {sonraki_satir_sayisi}")
    print("-" * 30)
```

```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re

path = kagglehub.dataset_download("mathurinache/list-of-countries-by-average-elevation")
csv_file = [f for f in os.listdir(path) if f.endswith('.csv')][0]
df = pd.read_csv(os.path.join(path, csv_file))


df.columns = df.columns.str.strip()
country_col = df.columns[0]
elevation_col = [c for c in df.columns if 'elevation' in c.lower()][0]


def clean_elevation(val):
    if pd.isna(val): return 0
    match = re.search(r'([-+]?[0-9,.]+)', str(val))
    if match:
        return float(match.group(1).replace(',', ''))
    return 0

df['elevation_numeric'] = df[elevation_col].apply(clean_elevation)


df_all = df.sort_values(by='elevation_numeric', ascending=True)


plt.figure(figsize=(12, 45))
sns.set_theme(style="whitegrid")


ax = sns.barplot(
    data=df_all,
    x='elevation_numeric',
    y=country_col,
    palette="viridis"
)


for i, v in enumerate(df_all['elevation_numeric']):
    ax.text(v + 5, i, f"{int(v)}m", va='center', fontsize=8)

plt.title('Tüm Ülkelerin Ortalama Yükseltileri (Sıralı Tam Liste)', fontsize=16)
plt.xlabel('Ortalama Yükselti (Metre)', fontsize=12)
plt.ylabel('Ülke', fontsize=10)


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Veriyi hazırlama
X = yearly_temp.index.values.reshape(-1, 1)
y = yearly_temp.values

# Train-test ayrımı test=%20
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# polinom regresyon özellikleri. 2. dereceden
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train)
X_test_poly = poly_features.transform(X_test)

# Eğitim
model = LinearRegression()
model.fit(X_train_poly, y_train)

# Gelecek Tahmini
future_years = np.arange(1850, 2101).reshape(-1, 1)
future_years_poly = poly_features.transform(future_years)
future_preds = model.predict(future_years_poly)

# Rmse hesabı
y_test_pred = model.predict(X_test_poly)
test_rmse = np.sqrt(mean_squared_error(y_test, y_test_pred))

# Görselleştirme
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

plt.scatter(X_train, y_train, color='blue', alpha=0.3, s=20, label='Eğitim Verisi (%80)')
plt.scatter(X_test, y_test, color='darkblue', alpha=0.6, s=25, label='Test Verisi (%20)')

plt.plot(future_years, future_preds, color='red', linewidth=3, label='Polinom Trend (2100 Projeksiyonu)')

plt.text(0.05, 0.9, f'Test RMSE: {test_rmse:.4f} °C', transform=plt.gca().transAxes,
         fontsize=13, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, edgecolor='red'))

plt.axvline(x=X_test[0], color='black', linestyle='--', alpha=0.3) # Split çizgisi
plt.text(X_test[0], plt.ylim()[0]+0.1, ' Test Ayrımı', fontsize=10)

plt.title('Polinom Regresyon ile Küresel Sıcaklık Projeksiyonu (2100)', fontsize=16, fontweight='bold')
plt.xlabel('Yıl')
plt.ylabel('Sıcaklık (°C)')
plt.legend(loc='lower right')
plt.show()

print(f"2100 Yılı Tahmini: {future_preds[-1]:.2f} °C")
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi hazırlama
df_xgb = yearly_temp.to_frame(name='temp')
df_xgb['year'] = df_xgb.index
df_xgb['diff'] = df_xgb['temp'].diff() # Yıllık artış miktarı

for i in range(1, 4):
    df_xgb[f'lag_{i}'] = df_xgb['diff'].shift(i)

df_xgb = df_xgb.dropna()

# Train Test ayrımı.
X = df_xgb.drop(columns=['temp', 'diff', 'year'])
y = df_xgb['diff']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# XGBoost modelini eğitme
model_xgb = xgb.XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=5)
model_xgb.fit(X_train, y_train)

# Hata hesaplama RMSE
y_pred_test = model_xgb.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

# Gelecek tahmini
future_years = np.arange(df_xgb['year'].max() + 1, 2101)
last_temp = df_xgb['temp'].iloc[-1]
current_lags = df_xgb.iloc[-1][['diff', 'lag_1', 'lag_2']].values

future_preds = []
temp_acc = last_temp

for yr in future_years:
    pred_diff = model_xgb.predict(current_lags.reshape(1, -1))[0]
    temp_acc += pred_diff
    future_preds.append(temp_acc)
    current_lags = np.array([pred_diff, current_lags[0], current_lags[1]])

# Görselleştirme
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

plt.plot(df_xgb['year'], df_xgb['temp'], color='#2c3e50', label='Gerçek Veri', alpha=0.7)

plt.plot(future_years, future_preds, color='#d35400', linewidth=3, linestyle='--', label='XGBoost 2100 Tahmini')

plt.text(0.05, 0.9, f'Test RMSE (Değişim): {rmse_test:.4f}', transform=plt.gca().transAxes,
         fontsize=12, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, edgecolor='#d35400'))

plt.title('XGBoost ile Küresel Sıcaklık Değişimi ve 2100 Projeksiyonu', fontsize=16)
plt.ylabel('Sıcaklık (°C)')
plt.legend()
plt.show()

print(f"XGBoost 2100 Yılı Tahmini: {future_preds[-1]:.2f} °C")
```

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Veri Hazırlama
full_path = os.path.join(path, 'Global_sea_level_rise.csv')
df_sea = pd.read_csv(full_path)
df_sea['date'] = pd.to_datetime(df_sea['date'])
df_sea['year'] = df_sea['date'].dt.year
yearly_sea = df_sea.groupby('year')['mmfrom1993-2008average'].mean().to_frame(name='level')

df_xgb = yearly_sea.copy()
df_xgb['diff'] = df_xgb['level'].diff()

for i in range(1, 4):
    df_xgb[f'lag_{i}'] = df_xgb['diff'].shift(i)

df_xgb = df_xgb.dropna()

# Train-Test ayrımı
X = df_xgb.drop(columns=['level', 'diff'])
y = df_xgb['diff']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# XGBoost modelini eğitme
model_xgb = xgb.XGBRegressor(n_estimators=150, learning_rate=0.05, max_depth=4)
model_xgb.fit(X_train, y_train)

# Gelecek Tahmini
future_years = np.arange(df_xgb.index.max() + 1, 2101)
last_real_level = df_xgb['level'].iloc[-1]
current_lags = df_xgb.iloc[-1][['diff', 'lag_1', 'lag_2']].values

future_preds = []
level_accumulator = last_real_level

for yr in future_years:
    pred_diff = model_xgb.predict(current_lags.reshape(1, -1))[0]
    level_accumulator += pred_diff
    future_preds.append(level_accumulator)

    current_lags = np.array([pred_diff, current_lags[0], current_lags[1]])

# Görselleştirme
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

plt.plot(yearly_sea.index, yearly_sea['level'], color='#2c3e50', label='Gerçek Veri', alpha=0.7)

plt.plot(future_years, future_preds, color='#e67e22', linewidth=3, linestyle='--', label='XGBoost 2100 Tahmini')

test_preds_diff = model_xgb.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, test_preds_diff))

plt.text(0.05, 0.9, f'Test RMSE (Değişim): {rmse_test:.4f} mm', transform=plt.gca().transAxes,
         fontsize=12, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, edgecolor='#e67e22'))

plt.title('XGBoost ile Deniz Seviyesi Artış Projeksiyonu (2100)', fontsize=16)
plt.ylabel('Deniz Seviyesi Değişimi (mm)')
plt.legend()
plt.show()

print(f"XGBoost 2100 Yılı Tahmini: {future_preds[-1]:.2f} mm")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Veriyi hazırlama
full_path = os.path.join(path, 'Global_sea_level_rise.csv')
df_sea = pd.read_csv(full_path)
df_sea['date'] = pd.to_datetime(df_sea['date'])
df_sea['year'] = df_sea['date'].dt.year
yearly_sea = df_sea.groupby('year')['mmfrom1993-2008average'].mean().to_frame(name='level')

df_rf = yearly_sea.copy()
df_rf['diff'] = df_rf['level'].diff()

for i in range(1, 4):
    df_rf[f'lag_{i}'] = df_rf['diff'].shift(i)

df_rf = df_rf.dropna()

# Train-Test ayrımı
X = df_rf.drop(columns=['level', 'diff'])
y = df_rf['diff']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Random forest model eğitimi
rf_model_sea = RandomForestRegressor(n_estimators=500, max_depth=10, random_state=42)
rf_model_sea.fit(X_train, y_train)

# Geleceği Tahmin Etme
future_years = np.arange(df_rf.index.max() + 1, 2101)
last_real_level = df_rf['level'].iloc[-1]
current_lags = df_rf.iloc[-1][['diff', 'lag_1', 'lag_2']].values

rf_predictions = []
level_acc = last_real_level

for yr in future_years:
    pred_diff = rf_model_sea.predict(current_lags.reshape(1, -1))[0]
    level_acc += pred_diff
    rf_predictions.append(level_acc)

    current_lags = np.array([pred_diff, current_lags[0], current_lags[1]])

# Hata hesaplama RMSE
test_preds_diff = rf_model_sea.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, test_preds_diff))

# Görselleştirme
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

plt.plot(yearly_sea.index, yearly_sea['level'], color='#34495e', label='Gerçek Veri', alpha=0.7)
plt.plot(future_years, rf_predictions, color='#27ae60', linewidth=3, linestyle='--', label='Random Forest 2100 Tahmini')

plt.text(0.05, 0.9, f'Test RMSE (Değişim): {rmse_test:.4f} mm', transform=plt.gca().transAxes,
         fontsize=12, fontweight='bold', bbox=dict(facecolor='white', alpha=0.8, edgecolor='#27ae60'))

plt.title('Random Forest ile Deniz Seviyesi Artış Analizi (2100)', fontsize=16)
plt.ylabel('Deniz Seviyesi Değişimi (mm)')
plt.legend()
plt.show()

print(f"Random Forest 2100 Yılı Tahmini: {rf_predictions[-1]:.2f} mm")
```
